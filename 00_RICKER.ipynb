# Ricker

```python
def ricker(f,length=0.256,dt=0.001):
    t=np.linspace(-length/2, (length-dt)/2, int(length/dt))
    y=(1.-2.*(np.pi**2)*(f**2)*(t**2))*np.exp(-(np.pi**2)*(f**2)*(t**2))
    return t,y
f=20
t,y=ricker(f)

plt.plot(t,y)
plt.title("Ricker wavelet, "+str(f)+"Hz")
```

## Add a slider

```python
length=0.256;
dt=0.001;

def ricker(freq):
    t=np.linspace(-length/2, (length-dt)/2, int(length/dt))
    y=(1.-2.*(np.pi**2)*(freq**2)*(t**2))*np.exp(-(np.pi**2)*(freq**2)*(t**2))
    plt.plot(t,y,c="green")
    plt.xlabel("Time")
    plt.ylabel("Amplitude")
    plt.title("Ricker wavelet")
# Interactivity
interact(ricker, freq=widgets.FloatSlider(value=40,
        min=10.0,
        max=80.0,
        step=5))

    
```
